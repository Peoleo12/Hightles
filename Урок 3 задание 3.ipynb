```python
r = 5
a = 2
if r >= 0 and 2*np.pi >= a >= 0:
    x = r*np.cos(a)
    y = r*np.sin(a)
    print('X =', x)
    print('Y =', y)
else:
    print('Ошибка ввода')
```
