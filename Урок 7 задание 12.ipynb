```python
def Q(x, y, z):
    return (x**2 + y**2 + z**2)
```

```python
# №1
x = np.linspace(-5, 5, 201)
y = 10 * x - 14
z = x + 2 * y - 1

plt.plot(x, Q(x, y, z))
plt.xlabel('x')
plt.ylabel('Q(x, y, z)')
plt.grid(True)
plt.show()
```

```python
# №2
x = np.linspace(0, 2, 201)
y = 10 * x - 14
z = x + 2 * y - 1

plt.plot(x, Q(x, y, z))
plt.xlabel('x')
plt.ylabel('Q(x, y, z)')
plt.grid(True)
plt.show()
```
