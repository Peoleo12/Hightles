```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = []
y1 = []
y2 = []
r = 1
for i in range(-100, 101):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt(r**2-xx**2))
    y2.append(-np.sqrt(r**2-xx**2))
    
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
x = []
y1 = []
y2 = []
a = 3
b = 2
for i in range(-300, 301):
    xx = i/100
    x.append(xx)
    y1.append(np.sqrt((1-xx**2/a**2)*b**2))
    y2.append(-np.sqrt((1-xx**2/a**2)*b**2))
    
plt.axis([-4.5, 4.5, -4.5, 4.5])
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
y = []
x1 = []
x2 = []
a = 1
b = 1
for i in range(-300, 301):
    yy = i/100
    #if xx <= -a or xx >= a:
    y.append(yy)
    x1.append(np.sqrt((1+yy**2/b**2)/a**2))
    x2.append(-np.sqrt((1+yy**2/b**2)/a**2))
    
plt.axis([-5.5, 5.5, -5.5, 5.5])
plt.plot(x1, y)
plt.plot(x2, y)
plt.xlabel('X')
plt.ylabel('Y')
```
